```python
import pandas as pd
```

```python
spy_data = "Resources/SPY_Data.xlsx"
msft_data = "Resources/MSFT.csv"
```

```python
spy_data_df = pd.read_excel(spy_data)
msft_data_df = pd.read_csv(msft_data)
```

```python
spy_data_df
```

```python
bins = [0, 50000000, 65000000, 80000000, 105000000, 140000000, 175000000, 250000000]

group_labels = ["< 50 mil", "50 - 65", "65 - 80", "80 - 105", "105 - 140", "140 - 175", "175 - 250"]

spy_data_df["Range"] = pd.cut(spy_data_df["Volume"], bins, labels=group_labels)

spy_data_df["%Gain"] = (spy_data_df["Close"] - spy_data_df["Open"]) / spy_data_df["Open"] *100

spy_data_df["Range of the day"] = spy_data_df["High"] - spy_data_df["Low"]

spy_data_df = spy_data_df.dropna()

spy_data_df
```

```python
spy_data_df.drop(spy_data_df.index[-1])
```

```python
volume_data = spy_data_df.groupby("Range")

volume_data["Volume"].describe()
```

```python
volume_data = spy_data_df.groupby("Range")

volume_data["%Gain"].describe()
```

```python
volume_data = spy_data_df.groupby("Range")

volume_data["Gain Next Day"].describe()
```

```python
volume_data = spy_data_df.groupby("Range")

volume_data["Range of the day"].describe()
```
